### Gradient Descent Alogrithm (Optimization Algorithm)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
# this is our function
f = lambda a : 12*a**2+7*a-4
# derivative function 
f_d = lambda a :24*a+7
```

### Ok! Let's plot the function

```python
x = np.linspace(-1,1,200)
y = [ f(a) for a in x]
plt.plot(x,y)
```

### lets minimize the function using gradient descent algorithm

```python
initialX = x[np.random.randint(0,100)]
# Learning Rate
eta = 0.01


def update(x_before):
    x_after = x_before - eta*f_d(x_before)
    return x_after


flag = True
while(flag):
    x_after = update(initialX)
    if abs(x_after - initialX) > 0.0001 :
        initialX = x_after
    else :
        initialX = x_after
        flag = False

# Plotting the minimum point 
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f (x)')
plt.scatter(initialX,f(initialX) , color='red')
print(initialX)
        
```

### It means if x = -0.2919786469971233 the output of function f, will be the minimum value.
